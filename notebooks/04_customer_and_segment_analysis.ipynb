# 04 Customer and Segment Analysis

Business focus: identify high-value customer groups, segment profitability, and regional segment opportunities.

```python
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(Path('../data/processed/global_superstore_cleaned.csv'))
df.head()
```

```python
segment = df.groupby('Segment', as_index=False).agg({'Sales': 'sum', 'Profit': 'sum', 'Customer_ID': 'nunique'})
segment['Profit_Margin'] = segment['Profit'] / segment['Sales']
segment.sort_values('Sales', ascending=False)
```

```python
sns.barplot(data=segment.sort_values('Sales', ascending=False), x='Sales', y='Segment')
plt.title('Sales by Customer Segment')
plt.show()
```

```python
region_segment = df.groupby(['Region', 'Segment'], as_index=False).agg({'Sales': 'sum', 'Profit': 'sum'})
region_segment['Profit_Margin'] = region_segment['Profit'] / region_segment['Sales']
region_segment.sort_values('Sales', ascending=False)
```

```python
top_customers = df.groupby(['Customer_ID', 'Customer_Name', 'Segment'], as_index=False).agg({'Sales': 'sum', 'Profit': 'sum', 'Order_ID': 'nunique'})
top_customers.sort_values('Sales', ascending=False).head(20)
```
