# 05 Profitability and Loss Analysis

Business focus: find high-sales low-margin areas, loss-making products, and discount patterns that reduce profit.

```python
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(Path('../data/processed/global_superstore_cleaned.csv'))
df.head()
```

```python
subcat = df.groupby('Sub_Category', as_index=False).agg({'Sales': 'sum', 'Profit': 'sum'})
subcat['Profit_Margin'] = subcat['Profit'] / subcat['Sales']
subcat.sort_values('Profit_Margin')
```

```python
high_sales_low_profit = subcat[(subcat['Sales'] > subcat['Sales'].median()) & (subcat['Profit_Margin'] < subcat['Profit_Margin'].median())]
high_sales_low_profit.sort_values('Profit_Margin')
```

```python
discount_profit = df.copy()
discount_profit['Discount_Band'] = pd.cut(discount_profit['Discount'], bins=[-0.01, 0, 0.1, 0.2, 0.3, 1], labels=['No discount', '0-10%', '11-20%', '21-30%', 'Above 30%'])
discount_profit.groupby('Discount_Band', observed=True).agg({'Sales': 'sum', 'Profit': 'sum'}).assign(Profit_Margin=lambda x: x['Profit'] / x['Sales'])
```

```python
negative_products = df.groupby('Product_Name', as_index=False).agg({'Sales': 'sum', 'Profit': 'sum'}).assign(Profit_Margin=lambda x: x['Profit'] / x['Sales'])
negative_products[negative_products['Profit'] < 0].sort_values('Profit').head(20)
```

## Interpretation

- Which products sell well but destroy margin?
- Does discounting increase revenue at the cost of profit?
- Which regions have high sales but weak profitability?
- What pricing, discount, or cost actions can management take?
