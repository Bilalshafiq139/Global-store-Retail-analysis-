# 03 Business Performance Analysis

Business focus: review KPIs, category performance, product performance, regional trends, and monthly sales movement.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

DATA_PATH = Path('../data/processed/global_superstore_cleaned.csv')
df = pd.read_csv(DATA_PATH, parse_dates=['Order_Date', 'Ship_Date'])
df.head()
```

```python
kpis = pd.Series({
    'Total Sales': df['Sales'].sum(),
    'Total Profit': df['Profit'].sum(),
    'Total Quantity': df['Quantity'].sum(),
    'Profit Margin': df['Profit'].sum() / df['Sales'].sum(),
    'Total Orders': df['Order_ID'].nunique(),
    'Total Customers': df['Customer_ID'].nunique()
})
kpis
```

```python
category_sales = df.groupby('Category', as_index=False)['Sales'].sum().sort_values('Sales', ascending=False)
sns.barplot(data=category_sales, x='Sales', y='Category')
plt.title('Sales by Category')
plt.show()
```

```python
top_products = df.groupby('Product_Name', as_index=False).agg({'Sales': 'sum', 'Profit': 'sum'}).sort_values('Sales', ascending=False).head(10)
top_products
```

```python
loss_products = df.groupby('Product_Name', as_index=False).agg({'Sales': 'sum', 'Profit': 'sum'}).sort_values('Profit').head(10)
loss_products
```

```python
monthly = df.groupby(['Year', 'Month'], as_index=False).agg({'Sales': 'sum', 'Profit': 'sum'})
sns.lineplot(data=monthly, x='Month', y='Sales', hue='Year', marker='o')
plt.title('Monthly Sales Trend')
plt.show()
```
